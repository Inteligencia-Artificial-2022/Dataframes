```python
import pandas as pd
```

```python
df = pd.read_csv("../data_sources/data_merged.csv")
```

```python
df
```

```python
dfnew = pd.DataFrame()
for y in df.columns[ 5: ]:
    dftemp = df.loc[ : , ['Country Name', 'Country Code', 'Continente', 'Sub-region', y]]
    dftemp["Year"] = y
    dftemp.columns = ['Country Name','Country Code', 'Continente', 'Sub-region','Value', 'Year']
    dftemp = dftemp.loc[ : , ['Country Name', 'Country Code', 'Continente', 'Sub-region','Year', 'Value']]
    dfnew = pd.concat([dfnew,dftemp])
    
```
