```python
import pandas as pd
df_data_merged = pd.read_csv("../data_sources/data_merged.csv", sep=";")
df_data_merged.head(1)
```

```python
#Eliminaci√≥n columnas innecesarias
df_data_cleaning = df_data_merged.drop(['id_asignatura','dni_catedratico', 'fecha_registro','fecha_nacimiento_x','genero_x','id_departamento','fecha_nacimiento_y','genero_y','id_carrera'], axis=1)
```

```python
df_data_cleaning
```

```python
#Reordenar las columnas
df_data_cleaning = df_data_cleaning.reindex(columns= ['dni_alumno','nombre_carrera','nombre_alumno','nombre_asignatura','nombre_departamento','nombre_catedratico','periodo','anio','promedio']) 
```

```python
#Renombrar las columnas del dataframe
df_data_cleaning = df_data_cleaning.rename(columns={'nombre_carrera':'carrera', 
                                                    'nombre_asignatura':'asignatura',
                                                    'nombre_alumno': 'alumno',
                                                    'nombre_catedratico': 'catedratico',
                                                    'nombre_departamento': 'departamento'
                                                   })
df_data_cleaning
```

```python
df_data_cleaning.to_csv('../data_sources/cleaned_data.csv', index = False, sep = ";")
```
